```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset

df = pd.read_csv("../data/raw_data.csv")

df.columns = df.columns.str.lower().str.strip().str.replace("-", "_", regex=False)

for col in df.select_dtypes(include=["object", "string"]).columns:
    df[col] = df[col].str.lower().str.strip().str.replace("-", "_", regex=False)



df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

Missing values were not properly detected, as some columns contained "?".
These were converted to NaN for accurate analysis.

```python
df = df.replace("?", pd.NA)

df.isnull().sum()
```

Missing values were identified in the categorical columns `workclass`, `occupation`, and `native_country`.

Instead of removing these rows, the missing values were filled using the label **"Unknown"**. This approach preserves the dataset while clearly indicating entries where information was not available.


```python
df["workclass"] = df["workclass"].fillna("unknown")
df["occupation"] = df["occupation"].fillna("unknown")
df["native_country"] = df["native_country"].fillna("unknown")

df.isnull().sum()
```

Fixing data types

The 'income' column contained an extra character, which was removed before converting it to numeric values (0 and 1).

```python
# df['income'].isnull().sum()
# df['income'].unique()
df["income"]
```

```python
df["income"] = df["income"].astype(str)  # convert to string first
df["income"] = df["income"].str.strip().str.replace(".", "", regex=False)
df["income"] = df["income"].map({"<=50k": 0, ">50k": 1})
```

```python
df['income'].dtype
```

The target variable was converted from float to integer type to ensure a clean and appropriate representation of binary classes for modeling.

```python
df["income"] = df["income"].astype("Int64") ## convert to Int64

df["income"].dtype
```

```python
cat_col = df.select_dtypes(include=["object", "string"]).columns # categorical columns
```

Converted categorical columns to category dtype for better memory efficiency and structured data representation.

```python
df[cat_col] = df[cat_col].astype("category")
```
