```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("../data/cleaned_data.csv")
```

## Distribution plot of numerical features

```python
cols = ["age", "hours_per_week", "capital_gain", "capital_loss"]

plt.figure(figsize=(10,6))

for i, col in enumerate(cols):
    plt.subplot(2, 2, i+1)   # 👈 simple indexing
    sns.histplot(data[col], kde=True)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.tight_layout()
plt.show()
```

## Boxplot (Outlier Visualization)

```python
cols = ["age", "capital_gain", "capital_loss", "hours_per_week"]

plt.figure(figsize=(10,6))

for i, col in enumerate(cols):
    plt.subplot(2, 2, i+1)   # 👈 same simple logic
    sns.boxplot(x=data[col])
    plt.title(f"{col} (Outlier Detection)")
    plt.xlabel(col)

plt.suptitle("Outlier Analysis of Numerical Features", fontsize=14)
plt.tight_layout()
plt.show()
```
